```python
import json
import pandas as pd
from airsenal.framework.utils import *
from airsenal.framework.prediction_utils import check_absence
```

Make a dict of records that will define template for our dataframe

```python
results_dict = {"season": [], "team": [], "week": [], "opponent": [], "home_or_away": [], "player": [], "injured_or_suspended": [], "reason": [], "minutes": [], "xG":[], "xA":[], "xGC":[], "bps":[], "goals":[], "assists":[], "goals_conceded":[]}
```

```python
season = "2425"
teams = list_teams(season)
```

```python
def get_player_minutes(team, gameweek, season, results_dict):
    fixtures = get_fixtures_for_gameweek(gw, season=season)
    players = list_players(team=team, season=season, gameweek=gw)
    for fixture in fixtures:
        if fixture.home_team != team and fixture.away_team != team:
            continue
        if fixture.home_team == team:
            opponent = fixture.away_team
            home_or_away = "home"
        else:
            opponent = fixture.home_team
            home_or_away = "away"
        print(f"Doing fixture {fixture.home_team} {fixture.away_team}")
        for player in players:
            playerscore = get_player_scores(fixture, player)
            if playerscore:
                results_dict["team"].append(team)
                results_dict["week"].append(gameweek)
                results_dict["season"].append(season)
                results_dict["opponent"].append(opponent)
                results_dict["home_or_away"].append(home_or_away)
                results_dict["player"].append(player.name)
                results_dict["minutes"].append(playerscore.minutes)
                results_dict["goals"].append(playerscore.goals)
                results_dict["assists"].append(playerscore.assists)
                results_dict["goals_conceded"].append(playerscore.conceded)
                results_dict["bps"].append(playerscore.bps)
                results_dict["xG"].append(playerscore.expected_goals)
                results_dict["xA"].append(playerscore.expected_assists)
                results_dict["xGC"].append(playerscore.expected_goals_conceded)
                absent, reasons, details = check_absence(player, gameweek, season)
                if not absent:
                    results_dict["injured_or_suspended"].append(False)
                    results_dict["reason"].append("n/a")
                else:
                    results_dict["injured_or_suspended"].append(True)
                    results_dict["reason"].append(reasons[0])
                
    return results_dict
                
    
```

```python

for team in teams:
    for gw in range(1,39):
        results_dict= get_player_minutes(team['name'], gw, season, results_dict)

        
```

```python
for k in results_dict.keys():
    print(k, len(results_dict[k]))
```

```python
df = pd.DataFrame(results_dict)
df.to_csv("player_minutes_with_extra_columns_2425.csv")
```
